```python
import pandas_datareader as pdr
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
import math
import csv

import gurobipy as gp
from gurobipy import GRB
```

```python
def minimize_var(v, mu, cov, shortExists=False):
    #create a model
    m = gp.Model("quadratic")
    #create variables
    if shortExists:
        theta = m.addMVar(mu.shape[0], lb=0.0, ub=1.0, name='theta')
    else:
        theta = m.addMVar(mu.shape[0], lb=float('-inf'), ub=float('inf'), name='theta')

    #set objective
    obj = theta @ cov @ theta
    m.setObjective(obj)

    #add Constraint
    m.addConstr((theta @ mu)-v==0, 'c0')
    m.addConstr(theta.sum()-1==0, 'c1')

    m.optimize()
    return m
```

```python
def minimize_var_with_rf(v, rf_rate, mu, cov, shortExists=False):
    #create a model
    m = gp.Model("quadratic")
    #create variables
    if shortExists:
        theta = m.addMVar(mu.shape[0], lb=0.0, ub=1.0, name='theta')
    else:
        theta = m.addMVar(mu.shape[0], lb=float('-inf'), ub=float('inf'), name='theta')

    #set objective
    obj = theta @ cov @ theta
    m.setObjective(obj)

    #add Constraint
    m.addConstr( ((mu-rf_rate) @ theta) == v-rf_rate, 'c0' )

    m.optimize()
    return m
```
